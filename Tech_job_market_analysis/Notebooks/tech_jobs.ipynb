```python
import pandas as pd

df = pd.read_csv("tech_jobs_salaries.csv")

df.head()
df.shape
df.info()
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
df.columns
```

```python
df['salary'] = pd.to_numeric(df['salary_local_currency'], errors='coerce')
```

```python
df['salary_lpa'] = df['salary']/100000
```

```python
def clean_role(title):
    if 'data analyst' in title:
        return 'data analyst'
    elif 'data scientist' in title:
        return 'data scientist'
    elif 'engineer' in title:
        return 'data engineer'
    else:
        return 'other'

df['job_role_clean'] = df['job_title'].apply(clean_role)
```

```python
def clean_text_column(col):
    return (col.astype(str)
              .str.strip()
              .str.lower()
              .str.replace(r'[^\w\s]', '', regex=True)
              .str.replace(r'\s+', '_', regex=True))

# Apply to all object columns
for col in df.select_dtypes(include='object').columns:
    df[col] = clean_text_column(df[col])
```
